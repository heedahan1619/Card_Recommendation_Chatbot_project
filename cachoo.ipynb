# 카추 - 카드 상품 추천 챗봇

## 1. 프로젝트 개요

### 1-1. 프로젝트명
> #### **카추** 
- 카드 상품 추천 챗봇

### 1-2. 프로젝트 추진 배경
> #### 사용자
- 카드 발급을 희망하는 개인 또는 법인 사용자
> #### 운영자
- 추천 챗봇 사이트
> #### 상황
- 과거에는 카드 설계사의 수익에 도움이 되는 카드 상품 위주로 이루어지며, 그마저도 설계사의 역량에 좌우되는 경우가 많았습니다.
- 그러나 생성형 AI의 도입으로 인해 현재는, 자신의 소비 패턴 및 이전 카드 이용 내역을 고려하여 필요한 카드 상품을 직접 조회하고 추천 받을 수 있습니다.
- 카드 상품 조회 및 추천 서비스를 챗봇 형태로 제공함으로써 사용자의 편의성을 증대시킬 것을 기대합니다.

### 1-3. 프로젝트 목표
- 사용자가 카드 발급 시에 중요하게 생각하는 요소를 간편하게 챗봇에 입력할 수 있도록 합니다.
- 사용자가 입력한 정보를 기반으로, 개인의 요구와 선호에 가장 적합한 맞춤형 카드를 추천합니다.
- 이를 통해 사용자의 금융 생활을 더욱 편리하고 유익하게 만들어 줍니다.

### 1-4. 프로젝트의 고도화 작업후 최종 목표
> **추천 시스템과 챗봇 프로세스를 구축**하고 다양한 데이터를 활용하여, 카드 상품 뿐만 아니라 대출, 주식, 펀드 등의 금융 상품부터 영화, OTT 컨텐츠 및 커머스 상품 추천까지 **다양한 항목에 적용**할 수 있는 플랫폼을 구축하는 것입니다.

## 2. 프로젝트 관리를 위한 일정(WBS)
총 3개월을 프로젝트 기간으로 설정합니다.

1주의 기간으로 설정합니다.

> #### 웹 크롤링을 통한 카드 정보 데이터 수집
- ##### scrapy를 사용하여 카드 고릴라 사이트에 있는 카드 정보 데이터 수집
    - 동적 구조로 이루어져 있어 json 파일을 불러오기
    - 필요한 내용만을 저장하고 정제하여 프로젝트에 활용할 수 있는 형태로 가공
    - json 형태로 저장
